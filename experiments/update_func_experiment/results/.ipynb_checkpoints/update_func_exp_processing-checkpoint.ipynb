```python
# optimal human
i = 0
oranges = ["#ffe9a3", "#f6de4b", "#ffa213", "#ec6d18"]
blues = ["#eff3f8", "#b2c4df", "#6689bf", "#3d5272"]
c = {}

for DEMPREF in [True, False]:
    plt.figure(i)
    plt.title("DEMPREF = " + str(DEMPREF))
    df = pd.DataFrame(np.zeros((0,5)), columns=["run #", "pref_iter", "type", "value", "update/n_query/inc_prev"])
    for UPDATE_FUNC in ["pick_best", "approx"]:
        for N_QUERY in [2, 3]:
            for INC_PREV_QUERY in [True, False]:
                try:
                    if not DEMPREF and INC_PREV_QUERY:
                        continue
                    if UPDATE_FUNC == "approx" and N_QUERY == 3:
                        continue
                    name = "dempref=" + str(DEMPREF) + ",inc=,n_query=,update=" + str(INC_PREV_QUERY) + "," + str(N_QUERY) + "," + str(UPDATE_FUNC) + "_db.pickle"
                    db = pickle.load(open(name, 'rb'), encoding='bytes')
                    label = "update=" + str(UPDATE_FUNC) + ",n_query=" + str(N_QUERY) + ",inc=" + str(INC_PREV_QUERY)
                    db["update/n_query/inc_prev"] = label
                    df = df.append(db, ignore_index=True)
                except:
                    continue
    m = df[df.type=='m']
    sns.pointplot(x='pref_iter', y='value', hue="update/n_query/inc_prev", data=m, palette=oranges.extend(blues))
    plt.legend(loc=(1.04,0))
    i += 1
```
